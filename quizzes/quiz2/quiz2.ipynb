(d) Create a function with four inputs $f(y,x,b0,b1)$

<font size = "3">

$\qquad e2 = (y - b_0 - b_1 * x)^2$

- The function should return the mean of e2
- This is the mean-square error (mse)
- Call the function "fn_mse"
- Let $y$ be the column "Potential"
- Let $x$ be the column "Age" 
- Does <br>
$\qquad (b_0 = 80, b_1 = 0.4)$ or <br>
$\qquad (b_0 = 80, b_1 =-0.3)$ or <br>
$\qquad (b_0 = 80, b_1 = -0.5)$ <br>
produce the lowest value of mse?
- Print your results

````python
# Write your own code
def fn_mse(y,x,b0,b1):
    e2 = (y-b0-b1*x)**2
    return e2.mean()

print(fn_mse(y=fifa23_players_basic["Potential"],x=fifa23_players_basic["Age"],b0=80,b1=0.4))
print(fn_mse(y=fifa23_players_basic["Potential"],x=fifa23_players_basic["Age"],b0=80,b1=-0.3))
print(fn_mse(y=fifa23_players_basic["Potential"],x=fifa23_players_basic["Age"],b0=80,b1=-0.5))
print("b0=80 and b1=-0.3 produces the lowest value of mse")
````

(e) Split a dataset into subsets

<font size = "3">

- Create ``` list_status = [0,1]```
- Create ``` list_prob = [0.5, 0.5]```
- Create a new variable $n$ with the number of rows of the dataset
- Define a random seed that equals 42

- Use 
```python
    np.random.choice(list_status,
                     size = n,
                     p = list_prob)
```
$\qquad$ to create a new column called "status" <br>
$\qquad$ which randomly assigns each row to "0" or "1" <br>
$\qquad$ with equal probability
- Use .query() to subset all observations with status == 0.<br>
store it in a new dataset called "data0"
- Use .query() to subset all observations with status == 1.<br>
store it in a new dataset called "data1"



````python
# Write your own code
list_status = [0,1]
list_prob = [0.5, 0.5]
n=len(fifa23_players_basic)
np.random.seed(42)
fifa23_players_basic["status"] = np.random.choice(list_status,
                 size = n,
                 p = list_prob)

data0=fifa23_players_basic.query("status == 0")
data1=fifa23_players_basic.query("status == 1")
````

<font size = "5">

(f) Create two overlapping histogram plots

<font size = "3">

- This question requires the two datasets you created in (e)
- Display the distribution of "Age" for "data0" and "data1"
- Use the "alpha" option to make the graphs semitransparent
- You will need to add a legend, label the axes, and the title

Note: The goal of this question is to illustrate that random <br>
assignment produces very similar distributions between two groups

- Provide a short interpretation of the graph using the information from the Note

````python
# Write your own code
data0["Age"].hist(alpha=0.2,label="Distribution of Ages for Players with Status 0")
data1["Age"].hist(alpha=0.2,label="Distribution of Ages for Players with Status 1")
plt.legend()
plt.title("Distribution of Ages for Fifa 2023 Players Based on Randomly Assigned Status 0 or 1")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
````

Given the graph above, the distribution of ages for players with status 0 and status 1 is very similar as the status was randomly assigned. There is no other factor contributing to any difference between the groups. The players come from the same overall data set and were randomly assigned to have status 1 or 0 causing the distribution of ages to be very similar.
